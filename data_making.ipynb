```python
%reset -f
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import math
from matplotlib.colors import LogNorm
from sklearn import mixture
```

```python
n_samples =200

# generate spherical data centered on (9, 9)
shifted_gaussian = 2*np.random.randn(n_samples,2) + np.array([9,9])

# generate zero centered stretched Gaussian data
C = np.array([[0., -0.7], [3.5, .7]])
stretched_gaussian = np.dot(np.random.randn(n_samples,2),C)

# generate (0,10) centered stretched Gaussian data
C = np.array([[0., -0.7], [3.5, 0]])
gaussian = np.dot(np.random.randn(n_samples,2),C) + np.array([-2,7])

# concatenate the two datasets into the final training set
X_train = np.vstack([shifted_gaussian, stretched_gaussian, gaussian])
```

```python
df = pd.DataFrame(X_train)

plt.scatter(X_train[:,0], X_train[:, 1], s=30, cmap='viridis');
```

```python
df.to_csv('three_normal.csv', index=False)
```
